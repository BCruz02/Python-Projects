```python
# This is an alarm clock created by Python!

import time 
import datetime
import pygame

def set_alarm(alarm_time):
    print(f"Alarm set for {alarm_time}")
    sound_file = r"C:\Users\Bcbra\PythonProject\SmallPythonProjects.py\SongForAlarmClock.mp3"
    is_running = True

    while is_running:
        current_time = datetime.datetime.now().strftime("%H:%M:%S")
        print(current_time)

        if current_time == alarm_input:
            print("TIME TO WAKE UP!!")

            pygame.mixer.init()
            pygame.mixer.music.load(sound_file)
            pygame.mixer.music.play()
            pygame.mixer.music.fadeout(15000)
            

            is_running = False

        time.sleep(1)
    

if __name__ == "__main__":
        
    while True:
        try:
            alarm_input = input("Enter the alarm time (HH:MM:SS)")
            datetime.datetime.strptime(alarm_input, "%H:%M:%S")
            set_alarm(alarm_input)
            break
        
        except ValueError:
            print("Invalid time format. Please use HH:MM:SS")
```
